```python
import os
base_dir = 'data/KDEF_and_AKDEF/KDEF/'
datas = []
for i, j, k in os.walk(base_dir):
#     print(k)
    for img in k:
        name = img.split('.')[0]
        No = name[:4]
        expression = name[4:6]
        angle = name[6:]
        swtich_expression = {
            "AF": 0,
            "AN": 1,
            "DI": 2,
            "HA": 3,
            "NE": 4,
            "SA": 5,
            "SU": 6
        }
        switch_angle = {
            "FL": 0,
            "FR": 1,
            "HL": 2,
            "HR": 3,
            "S" : 4
        }
        data =  No + '/' + img + ',' + \
                str(swtich_expression[expression]) + ',' + \
                str(switch_angle[angle])
        datas.append(data)
        print(data)
```

```python
import numpy as np
# 
num_test_samples = 1000
num_validation_samples = 1000
```

```python
# 
def save(fileName, contents):
    fh = open(fileName, 'w', encoding='utf-8')
    for item in contents:
        fh.write(str(item)+"\n")
    fh.close()

save('data/all.txt', datas)
```

```python
# 
np.random.shuffle(datas) # 

validation_data = datas[:num_test_samples]
test_data = datas[num_test_samples: num_test_samples+num_validation_samples]
training_data = datas[num_test_samples+num_validation_samples:]
len(validation_data), len(test_data), len(training_data)
```

```python
save('data/validation.txt', validation_data)
save('data/test.txt', test_data)
save('data/train.txt', training_data)
```
